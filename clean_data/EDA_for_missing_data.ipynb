Checar valores nulos en las variables importantes que usaremos para entrenar el modelo

```python
import pandas as pd

# Cargar archivos
games = pd.read_csv("../data/Games.csv", parse_dates=["gameDate"])
team_stats = pd.read_csv("../data/TeamStatistics.csv", parse_dates=["gameDate"])

# 1️⃣ Revisar columnas obligatorias
print("Games.csv columns:", games.columns.tolist())
print("TeamStatistics.csv columns:", team_stats.columns.tolist())

# 2️⃣ Revisar nulos
print("\nNulos en Games.csv:\n", games.isnull().sum())
print("\nNulos en TeamStatistics.csv:\n", team_stats.isnull().sum())

# 3️⃣ Revisar tipos de dato
print("\nGames.csv dtypes:\n", games.dtypes)
print("\nTeamStatistics.csv dtypes:\n", team_stats.dtypes)

# 4️⃣ Revisar valores únicos de equipos (consistencia)
print("\nEquipos en Games.csv (home):", games["hometeamName"].unique())
print("Equipos en TeamStatistics.csv:", team_stats["teamName"].unique())

# 5️⃣ Revisar temporadas ≥ 2015
games_2015 = games[games["gameDate"].dt.year >= 2015]
print("\nTotal juegos desde 2015:", len(games_2015))
print("Año mínimo:", games_2015["gameDate"].dt.year.min())
print("Año máximo:", games_2015["gameDate"].dt.year.max())

# 6️⃣ Revisar que scores sean numéricos y no nulos
print("\nGames scores info:\n", games[["homeScore","awayScore"]].describe())
print("TeamStatistics teamScore/opponentScore info:\n", team_stats[["teamScore","opponentScore"]].describe())
```
